```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("book.csv",encoding='latin1')
df.head()
```

```python
df=df.rename(columns={'Unnamed: 0':'Index','User.ID':'UserID','Book.Title':'BookTitle','Book.Rating':'Rating'})
```

```python
df.info()
```

```python
df.drop('Index',axis=1,inplace=True)
```

```python
len(df.BookTitle.unique())
```

```python
user_df = df.groupby(['UserID', 'BookTitle'])['Rating'].mean().unstack()
```

```python
user_df = df.pivot_table(index='UserID', columns='BookTitle', values='Rating', aggfunc='mean')
user_df
```

```python
user_df.fillna(0, inplace=True)
user_df
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
user_sim = 1-pairwise_distances(user_df.values,metric='cosine')
```

```python
user_sim
```

```python
user_sim.shape
```

```python
user_sim_df=pd.DataFrame(user_sim)
user_sim_df
```

```python
user_sim_df.index=df.UserID.unique()
user_sim_df.columns=df.UserID.unique()
```

```python
user_sim_df.iloc[0:15, 0:15]
```

```python
np.fill_diagonal(user_sim, 0)
user_sim_df.iloc[0:15, 0:15]
```
